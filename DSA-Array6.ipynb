```python
#Q1
class Solution {
public:
    vector<int> diStringMatch(string s) {
        int n = s.size();
        int i = 0;
        int j = n;

        vector<int>ans;

        for(int k = 0; k<=n; k++){
            if(s[k]=='I'){
                ans.push_back(i);
                i++;
            }else{
                ans.push_back(j);
                j--;
            }
        }

        if(i == j){
            ans.push_back(i);
        }

        return ans;
    }
};
```

```python
#Q2
class Solution {
public:
    bool searchMatrix(vector<vector<int>>& arr, int target) {
        int n=arr.size();
        int m=arr[0].size();
        int i=0;
        
        while(i<n)
        {
        if(target<=arr[i][m-1])
        {
            int j=0;
            while(j<m)
            {
                if(target==arr[i][j])
                return true;
                else if(target<arr[i][j])
                return false;
                else
                j++;
            }
            return false;
        }
        i++;
        }
               
    return false;
    }
};
```

```python
#Q3
class Solution {
public:
    bool validMountainArray(vector<int>& arr) {
        if(arr.size()<3)
        return false;
        if(arr[0]>arr[1])
            return false;
        int i=0;
        while (i < arr.size()-1 && arr[i] < arr[i+1]) i++; 
        if (i == 0 || i == arr.size()-1)
            return false;
        while (i < arr.size()-1 && arr[i] > arr[i+1]) i++;
        return i == arr.size()-1;
    }
};
```

```python
#Q4
class Solution {
public:
    int findMaxLength(vector<int>& nums) {
        unordered_map<int,int> mp;
        int x=0;
        mp[0]=-1;
        int ans=0;
        for(int i=0;i<nums.size();i++)
        {
            if(nums[i]==0)
                x--;
            else
                x++;
            if(mp.find(x)!=mp.end())
                ans=max(ans,i-mp[x]);
            else
                mp[x]=i;
        }
        return ans;
    }
};
```

```python
#Q5
class Solution {
public:
    int minProductSum(vector<int>& A, vector<int>& B) {
        sort(begin(A), end(A));
        sort(begin(B), end(B), greater<>());
        int ans = 0;
        for (int i = 0; i < A.size(); ++i) ans += A[i] * B[i];
        return ans;
    }
};
```

```python
#Q6
class Solution {
public:
    vector<int> findOriginalArray(vector<int>& changed) {
        int n = changed.size();
        if (n % 2 == 1) return {};
        sort(changed.begin(), changed.end());
        vector<int> ans;
        map<int, int> mp;
        for (int i = 0; i < n; i++) {
            mp[changed[i]]++;
        }
        for (int i = 0; i < n; i++) {
          if (mp[changed[i]] == 0) continue;
          if (mp[changed[i] * 2] == 0) return {};
          ans.push_back(changed[i]);
          mp[changed[i]]--;
          mp[changed[i] * 2]--;
        }
        return ans;
    }
};
```
