```python
# 20. Implement a function to reverse the order of words in a given sentence.
def reverse_sentence(sentence):
    words = sentence.split()
    reversed_words = words[::-1]
    reversed_sentence = ' '.join(reversed_words)
    return reversed_sentence

# 21. Write a Python program to check if two strings are anagrams of each other.
def is_anagram(str1, str2):
    str1 = str1.lower()
    str2 = str2.lower()
    
    # Remove whitespace and punctuation
    str1 = ''.join(e for e in str1 if e.isalnum())
    str2 = ''.join(e for e in str2 if e.isalnum())
    
    # Check if the lengths are equal
    if len(str1) != len(str2):
        return False
    
    # Create dictionaries to store character counts
    char_count1 = {}
    char_count2 = {}
    
    # Update character counts for str1
    for char in str1:
        if char in char_count1:
            char_count1[char] += 1
        else:
            char_count1[char] = 1
    
    # Update character counts for str2
    for char in str2:
        if char in char_count2:
            char_count2[char] += 1
        else:
            char_count2[char] = 1
    
    # Compare character counts
    return char_count1 == char_count2

# 22. Implement a function to find the first non-repeating character in a string.
def first_non_repeating_character(string):
    char_count = {}

    for char in string:
        char_count[char] = char_count.get(char, 0) + 1

    for char in string:
        if char_count[char] == 1:
            return char

    return None

# 23. Write a program to find the prime factors of a given number.
def find_prime_factors(num):
    factors = []
    divisor = 2

    while divisor <= num:
        if num % divisor == 0:
            factors.append(divisor)
            num = num / divisor
        else:
            divisor += 1

    return factors

# 24. Implement a function to check if a given number is a power of two.
def is_power_of_two(num):
    if num <= 0:
        return False
    return (num & (num - 1)) == 0

# 25. Write a Python program to merge two sorted lists into a single sorted list.
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = 0
    j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    merged_list.extend(list1[i:])
    merged_list.extend(list2[j:])

    return merged_list
```

```python
# 26. Implement a function to find the mode of a list of numbers.
from collections import defaultdict

def find_mode(numbers):
    counts = defaultdict(int)

    for num in numbers:
        counts[num] += 1

    max_count = max(counts.values())
    mode = [num for num, count in counts.items() if count == max_count]

    return mode

# 27. Write a program to find the greatest common divisor (GCD) of two numbers.
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

# 28. Implement a function to calculate the square root of a given number.
def square_root(n):
    if n < 0:
        raise ValueError("Square root is not defined for negative numbers.")
    
    if n == 0 or n == 1:
        return n
    
    # Binary search bounds
    left = 1
    right = n
    
    while left <= right:
        mid = (left + right) // 2
        
        # Check if mid*mid is equal to n
        if mid * mid == n:
            return mid
        
        # If mid*mid is less than n, search in the right half
        if mid * mid < n:
            left = mid + 1
            result = mid
        
        # If mid*mid is greater than n, search in the left half
        else:
            right = mid - 1
    
    return result

# 29. Write a Python program to check if a given string is a valid palindrome ignoring non-alphanumeric characters.
def is_valid_palindrome(s):
    # Remove non-alphanumeric characters and convert to lowercase
    s = ''.join(char.lower() for char in s if char.isalnum())
    
    # Check if the string is equal to its reverse
    return s == s[::-1]

# 30. Implement a function to find the minimum element in a rotated sorted list.
def find_min(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        # Check if the mid element is greater than the rightmost element
        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]

# 31. Write a program to find the sum of all even numbers in a list.
def sum_even_numbers(nums):
    return sum(num for num in nums if num % 2 == 0)
```
